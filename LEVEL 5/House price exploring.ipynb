```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('housing_price_dataset.csv')
```

```python
summary_stats=df[['SquareFeet','Price']].describe()
```

```python
df['price_per_sqft']=df['Price']/df['SquareFeet']
```

```python
plt.scatter(df['SquareFeet'], df['Price'], alpha=0.5)
plt.title('Price vs. Square Footage')
plt.xlabel('Square Footage (sqft)')
plt.ylabel('Price ($)')
plt.grid(True)

print("First 5 rows of the updated DataFrame:")
print(df.head())

plt.show()
```
