```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
data=np.array([11,22,33,44,55,66,77,88,99,100])
```

```python
Q1=np.percentile(data,25)

Q3=np.percentile(data,75)
```

```python
IQR=Q3-Q1
```

```python
plt.boxplot(data)
plt.title("Box Plot of the Dataset")
plt.show()

print(f"IQR = {IQR}")
```
