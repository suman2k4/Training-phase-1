```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python

digits = load_digits()
X = digits.data  
y = digits.target
```

```python

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python

pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)
```

```python

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], 
                     c=y, cmap='Spectral', s=50, alpha=0.6)

legend = ax.legend(*scatter.legend_elements(), title="Digits")
ax.add_artist(legend)

ax.set_title("3D PCA of Digits Dataset")
ax.set_xlabel("PCA 1")
ax.set_ylabel("PCA 2")
ax.set_zlabel("PCA 3")
plt.show()
```

```python

explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
print(" Explained variance ratio (each component):", explained_variance)
print(" Cumulative explained variance (3 components):", cumulative_variance[-1])
```
