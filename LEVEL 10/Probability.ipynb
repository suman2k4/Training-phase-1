```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom
import numpy as np
```

```python
data=pd.read_csv('online_retail.csv')
```

```python
high_order_threshold = 50
data['High_Order'] = data['Quantity'] > high_order_threshold
high_order_count = data['High_Order'].sum()
total_orders = len(data)
probability = high_order_count / total_orders
print(f"Probability of High Orders: {probability}")
```

```python
n = total_orders
p = probability
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)
```

```python

plt.bar(x, pmf)
plt.title('Binomial Distribution of High Orders')
plt.xlabel('Number of High Orders')
plt.ylabel('Probability')
plt.show()
```
