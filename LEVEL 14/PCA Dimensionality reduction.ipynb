```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.datasets import load_wine
```

```python
data=load_wine()
```

```python
X=pd.DataFrame(data.data,columns=data.feature_names)
y=data.target
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_scaled)
```

```python
variance_ratio=pca.explained_variance_
```

```python
total_variance=sum(variance_ratio)
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', alpha=0.7)
plt.xlabel(f'PC1 ({variance_ratio[0]:.1%} variance)')
plt.ylabel(f'PC2 ({variance_ratio[1]:.1%} variance)')
plt.title(f'Wine Dataset PCA (Total Variance Explained: {total_variance:.1%})')
plt.colorbar(scatter, label='Wine Class')
plt.grid(alpha=0.3)
plt.show()
```

```python

print("Explained Variance Ratio:", variance_ratio)
print(f"Total Variance Explained: {total_variance:.1%}\n")

print("Interpretation:")
print(f"- The first two principal components capture {total_variance:.1%} of the total variance")
print("- PC1 accounts for most of the variance ({variance_ratio[0]:.1%})")
print("- The clear separation between classes in the plot shows that")
print("  these two components retain meaningful structure from the original 13 features")
```
