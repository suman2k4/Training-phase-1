```python
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python

X, _ = make_blobs(n_samples=300, centers=3, n_features=2, random_state=42)
X = StandardScaler().fit_transform(X)
```

```python

Z = linkage(X, method='ward')
```

```python

plt.figure(figsize=(10, 5))
dendrogram(Z, truncate_mode='lastp', p=12)
plt.title('Customer Segmentation Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python

clusters = fcluster(Z, t=3, criterion='maxclust')
```

```python

plt.figure(figsize=(8, 6))
plt.scatter(X[:,0], X[:,1], c=clusters, cmap='viridis', alpha=0.7)
plt.title('Customer Clusters (k=3)')
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.show()

print("The dendrogram shows three distinct clusters forming at distance ~5.")
print("This suggests natural grouping of customers into 3 segments.")
```
